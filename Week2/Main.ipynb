```python
import pandas as pd

# Load the dataset
url = "https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/wetter.csv"
df = pd.read_csv(url)

# Convert 'Datum' to datetime format
df['Datum'] = pd.to_datetime(df['Datum'])

# 1. Overall average temperature
overall_avg_temp = df['Temperatur'].mean()
print("Overall average temperature:", overall_avg_temp)

# 2. Average temperature in July
july_avg_temp = df[df['Datum'].dt.month == 7]['Temperatur'].mean()
print("Average temperature in July:", july_avg_temp)

# 3. Average temperature in May
may_avg_temp = df[df['Datum'].dt.month == 5]['Temperatur'].mean()
print("Average temperature in May:", may_avg_temp)

# 4. Comparison
if july_avg_temp > may_avg_temp:
    print("July is warmer than May on average.")
elif july_avg_temp < may_avg_temp:
    print("May is warmer than July on average.")
else:
    print("July and May have the same average temperature.")
```

```python
import pandas as pd

data = pd.read_csv("kiwo.csv")
import matplotlib.pyplot as plt

df = pd.DataFrame(data)
df['Datum'] = pd.to_datetime(df['Datum'])  # Convert string to datetime
df['Year'] = df['Datum'].dt.year           # Extract year for grouping

# Step 2: Plotting
plt.figure(figsize=(14, 10))

# Scatterplot
plt.subplot(3, 1, 1)
plt.scatter(df['Datum'], df['KielerWoche'], color='blue')
plt.title("Scatterplot: Kieler Woche")
plt.xlabel("Datum")
plt.ylabel("KielerWoche")

# Histogram
plt.subplot(3, 1, 2)
plt.hist(df['Year'], bins=range(df['Year'].min(), df['Year'].max() + 2), color='green', edgecolor='black', align='left')
plt.title("Histogramm: Anzahl der Kieler Woche Tage pro Jahr")
plt.xlabel("Jahr")
plt.ylabel("Anzahl")

# Bar Chart (Balkendiagramm)
plt.subplot(3, 1, 3)
year_counts = df['Year'].value_counts().sort_index()
plt.bar(year_counts.index.astype(str), year_counts.values, color='orange')
plt.title("Balkendiagramm: Kieler Woche Tage pro Jahr")
plt.xlabel("Jahr")
plt.ylabel("Anzahl")

plt.tight_layout()
plt.show()
```
